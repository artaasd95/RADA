# Download / verify Qwen portfolio models

Caches all models from `configs/models/qwen_portfolio.yaml` under `RADA_MODEL_CACHE_ROOT`.
Skips download when a valid local directory already exists.

```python
import os
from pathlib import Path

from rada.models import load_model_registry, resolve_model_path, get_cache_root

cache = get_cache_root()
print(f"RADA_MODEL_CACHE_ROOT -> {cache}")
registry = load_model_registry()
rows = []
for model_id in registry.list_model_ids():
    entry = registry.get_model(model_id)
    try:
        path = resolve_model_path(model_id, registry=registry, download=True)
        status = "found" if path.exists() else "missing"
    except Exception as exc:
        path = None
        status = f"error: {exc}"
    rows.append((model_id, entry.hub_path, status, str(path)))

for model_id, hub, status, path in rows:
    print(f"{model_id:24} {hub:35} {status:10} {path}")
```
